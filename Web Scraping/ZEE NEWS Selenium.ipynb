```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
```

## ZEE NEWS

```python
driver.get('https://zeenews.india.com/hindi/tags/nupur-sharma.html')
```

```python
for i in range(0, 5):
    try:
        driver.find_element(By.ID, "readmorenews").click()

    except NoSuchElementException:
        break
```

```python
news_list = driver.find_elements(By.CLASS_NAME, "news_item")
```

```python
news_num = len(news_list)
news_num
```

```python
all_news_list = []

for i in range(0, news_num):
    row_dict = {}
    # Extracting news title
    news_title1 = news_list[i].find_element(By.CLASS_NAME, "news_title")
    row_dict['news_title'] = news_title1.text

    # Extracting news description
    news_details1 = news_list[i].find_element(By.CLASS_NAME, "news_details")
    row_dict['news_details'] = news_details1.text

    # Extracting date
    news_date1 = news_list[i].find_element(By.CLASS_NAME, "news_updatetime")
    row_dict['news_date'] = news_date1.get_attribute('textContent')

    # Channel/Source Name
    row_dict['source_name'] = 'ZEE News'

    all_news_list.append(row_dict)
```

```python
data = pd.DataFrame(all_news_list)
```
