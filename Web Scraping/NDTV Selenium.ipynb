```python
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC

import pandas as pd
```

## NDTV NEWS (Site is NDTV India)

```python
driver.get('https://ndtv.in/topic/nupur-sharma')
```

```python
s = driver.find_element(By.ID, "lodmorenews")
driver.execute_script("arguments[0].click();",s)
```

```python
for i in range(0, 10):
    try:
        s = driver.find_element(By.ID, "lodmorenews")
        driver.execute_script("arguments[0].click();",s)

    except NoSuchElementException:  
        break
```

```python
news_list = driver.find_elements(By.CLASS_NAME, "src_lst-li")
print(len(news_list))
```

```python
try:
    elem = news_list[4].find_element(By.CLASS_NAME, "newsHdng")
except NoSuchElementException: 
    pass
```

```python
news_date1 = news_list[71].find_element(By.CLASS_NAME, "src_itm-stx")
news_date1.text.split('|')
```

```python
all_news_list_ndtv = []

for i in range(0, len(news_list)):
    row_dict = {}
    
    try:
        # Extracting news title
        news_title1 = news_list[i].find_element(By.CLASS_NAME, "src_itm-ttl")
        row_dict['news_title'] = news_title1.text

              # Extracting news description
        news_details1 = news_list[i].find_element(By.CLASS_NAME, "src_itm-txt")
        row_dict['news_details'] = news_details1.text

        # Extracting date
        news_date1 = news_list[i].find_element(By.CLASS_NAME, "src_itm-stx")
        
        info_list = news_date1.text.split('|')

        row_dict['news_date'] = info_list[len(info_list) - 1]

        # Channel/Source Name
        row_dict['source_name'] = 'NDTV News'   
        all_news_list_ndtv.append(row_dict)  

    except NoSuchElementException:  
        pass
```

```python
data = pd.DataFrame(all_news_list_ndtv)
```
