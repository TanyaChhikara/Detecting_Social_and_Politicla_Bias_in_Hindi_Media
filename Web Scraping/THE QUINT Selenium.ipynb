```python
full_dataset = []

for i in range(0, len(link)):
    data_set = {}
    try:
        driver.get(link[i])
        # Extracting news title
        title = driver.find_element(By.CLASS_NAME,'ieh5X.RXXC0')
        data_set['news_title'] = title.text

        # Extracting news description
        content = driver.find_element(By.CLASS_NAME,'_2QRqw')
        data_set['news_details'] = content.text

        # Extracting date
        news_date = driver.find_element(By.CLASS_NAME, '_1yYGD')
        data_set['news_date'] = news_date.text

        # Channel/Source Name
        data_set['source_name'] = 'The Quint'   
        full_dataset.append(data_set)  

    except NoSuchElementException:  
        pass
```

```python
data = pd.DataFrame(full_dataset)
```
